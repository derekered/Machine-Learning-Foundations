```python
import graphlab
graphlab.canvas.set_target('ipynb')
```

```python
products = graphlab.SFrame('amazon_baby.gl/')
```

```python
products['word_count'] = graphlab.text_analytics.count_words(products['review'])
```

```python
products.head()
```

# Put Select Words Into Array

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

# Create Function to Get Word Count

```python
def word_count(word, word_count):
    if word in word_count:
        return word_count[word]
    else:
        return 0
```

# Apply Function to Data Set

```python
for word in selected_words:
    products[word] = products['word_count'].apply(lambda x: word_count(word,x))
```

```python
products.head()
```

# Most Common Words

```python
common_words = {}
```

```python
for word in selected_words:
    common_words[word] = products[word].sum()

print common_words
```

# Define Sentiment

```python
# Ignore all 3* reviews
products = products[products['rating'] != 3]
```

```python
# Give 4 and 5 star reviews value of 1
products['sentiment'] = products['rating'] >=4
```

```python
products.head()
```

```python
products['sentiment'].mean()
```

# Train Sentiment Model with Selected Words

```python
train_data, test_data = products.random_split(.8, seed=0)
```

```python
sentiment_model = graphlab.logistic_classifier.create(train_data, target='sentiment', features=selected_words, validation_set=test_data)
```

# Evaluate Model

## Evaluate Coefficients

```python
coefficients = sentiment_model['coefficients']
```

```python
coefficients = coefficients.sort('value')
```

```python
coefficients.print_rows(12)
```

## Evaluate Accuracy

```python
sentiment_model.evaluate(test_data)
```

```python
sentiment_model.show(view='Evaluation')
```
